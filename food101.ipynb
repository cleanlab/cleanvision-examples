### Analysis of issues found per class

Here we break down the issues within different groups of images. Since this is a classification dataset, we group images that share the same class label (but note that CleanVision can be run on any collection of images, regardless if the ML task is classification or not).

Add `class name` column to `imagelab.issues` dataframe

```python
issues_df = imagelab.issues.copy()
class_column = "class name"
issues_df[class_column] = [path.split("/")[-2] for path in issues_df.index.to_list()]
```

Compute number of instances of each issue per class and also the num_images in each class

```python
issues_per_class = issues_df.groupby('class name').agg(
    num_odd_aspect_ratio=pd.NamedAgg(column="is_odd_aspect_ratio_issue", aggfunc="sum"),
    num_low_information=pd.NamedAgg(column="is_low_information_issue", aggfunc="sum"),
    num_light=pd.NamedAgg(column="is_light_issue", aggfunc="sum"),
    num_grayscale=pd.NamedAgg(column="is_grayscale_issue", aggfunc="sum"),
    num_dark=pd.NamedAgg(column="is_dark_issue", aggfunc="sum"),
    num_blurry=pd.NamedAgg(column="is_blurry_issue", aggfunc="sum"),
    num_exact_duplicates=pd.NamedAgg(column="is_exact_duplicates_issue", aggfunc="sum"),
    num_near_duplicates=pd.NamedAgg(column="is_near_duplicates_issue", aggfunc="sum"),
    num_images=pd.NamedAgg(column="class name", aggfunc="count")
)
```

```python
issues_per_class
```

Top classes with blurry issue

```python
issues_per_class.sort_values(by=['num_blurry'], ascending=False)[['num_blurry', 'num_images']].head()
```

Top classes with dark issue

```python
issues_per_class.sort_values(by=['num_dark'], ascending=False)[['num_dark', 'num_images']].head()
```

Top classes with exact duplicates issue

```python
issues_per_class.sort_values(by=['num_exact_duplicates'], ascending=False)[['num_exact_duplicates', 'num_images']].head()
```

Top classes with near duplicates issues

```python
issues_per_class.sort_values(by=['num_near_duplicates'], ascending=False)[['num_near_duplicates', 'num_images']].head()
```
